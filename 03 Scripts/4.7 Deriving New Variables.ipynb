# creating a flag to sort products

```python
# create subset of first 1000000 rows 
df=df_ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
# Define a function - price label
def price_label(row):
    
    if row['prices'] <=5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range products'
    elif row['prices'] >15:
        return 'High range'
    else: return 'Not enough data'
```

```python
#apply the function
df['price_range']= df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

```python
# checking for most expensive prodcut 
df['prices'].max()
```

# Procedures for Task

```python
# deleting of the first 'busiest day' column to avoid confusion 
df_ords_prods_merge=df_ords_prods_merge.drop(columns=['busiest_day'])
```

## loop for 2 top days and 2 lowest days updated

```python
#updated look for top 2 and lowest 2. adding the in operator
results = []

for value in df_ords_prods_merge['orders_day_of_week']:
    if value in [0, 1]:
        results.append('Busiest days')
    elif value in [3, 4]:
        results.append('Slowest days')
    else:
        results.append('Regularly busy')
```

```python
#assigning the list to a new column 
df_ords_prods_merge['busiest_days']= results
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna=False)
```
