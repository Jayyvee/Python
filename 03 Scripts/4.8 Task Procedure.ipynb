```python
# loyalty flag for customers with transform () and loc()
# create a loyalty flag using loc()- create ' max order'
ords_prods_merge['max_order']= ords_prods_merge.groupby (['user_id']) ['order_number'].transform ('max')
```

```python
ords_prods_merge.head(5)
```

## loyalty flag using loc() function


```python
#If the max orders the user has made is over 40, then the customer will be labeled a ‘loyal customer’
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag']= 'Loyal customer'
```

```python
#In the max orders the user has made is over but less than or equal to 40, then the customer will be labeled a ‘ regular customer’ 
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag']= 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag']= 'New customer'
```

```python
#checking frequency
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
# Replace prices > 100 with NaN // getting rid of outliers
ords_prods_merge.loc[ords_prods_merge['prices'] > 100, 'prices'] = np.nan
```

```python
# compare the prices for loyalty column
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean','min','max', 'std']})
```

# answer:
we are able to see that Loyal customers average spending to be $7.78 compared to New customers at $7.80 and Regualr customers at $7.80. new customers tend to spend slighting more than loyal customers 

# create a spending flag

```python
#aggregating data with transformed ()
ords_prods_merge['avg_product_price']= ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
ords_prods_merge[['user_id','avg_product_price', 'spending_flag', 'Median_days_between']].head(15)
```

```python
# Sort by user_id to see the grouping, and order_number to see the sequence
ords_prods_merge = ords_prods_merge.sort_values(by=['user_id', 'avg_product_price'])
```

```python
#if the mean of the price of prodcuts purchased by a user is lower than 10, then flag as a 'low spender'
ords_prods_merge.loc[ords_prods_merge['avg_product_price'] < 10, 'spending_flag']= 'Low spender'
```

```python
#if the mean of the prices of products purchased by a user is higher than or equal to 10, then flag as 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_product_price'] >= 10, 'spending_flag']= 'High spender'
```

##  Frequent customers column

```python
#aggregating data with transformed ()
ords_prods_merge['Median_days_between']= ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# Sort by user_id to see the grouping, and order_number to see the sequence
ords_prods_merge = ords_prods_merge.sort_values(by=['user_id', 'days_since_prior_order'])
```

```python
# if the median of "days since prior order' is higher than 20, then the customer is 'non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['Median_days_between'] >20, 'frequent_flag']= 'non-frequent customer'
```

```python
#If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
ords_prods_merge.loc[(ords_prods_merge['Median_days_between'] >10) & (ords_prods_merge['frequent_cust'] <=20), 'frequent_flag']= 'Regular customer'
```

```python
#If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
ords_prods_merge.loc[ords_prods_merge['Median_days_between'] <=10, 'frequent_flag']= 'frequent customer'
```

```python
#checking frequency
ords_prods_merge['frequent_flag'].value_counts(dropna=False)
```
