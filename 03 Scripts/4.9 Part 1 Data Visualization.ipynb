# Data Wrangling

## rename columns

```python
# change column name ( surnam) to Last_name
df_cust.rename(columns={'Surnam':'Last_name'}, inplace=True)
```

```python
# change column name State
df_cust.rename(columns={'STATE':'State'}, inplace=True)
```

```python
# change column name n_dependants
df_cust.rename(columns={'n_dependants':'#_dependants'}, inplace=True)
```

```python
# change column name  First name
df_cust.rename(columns={'First Name':'First_name'}, inplace=True)
```

```python
# change column name ( surnam) to Last_name
df_cust.rename(columns={'fam_status':'marital_status'}, inplace=True)
```

### dropping columns

```python
#dropping frist and last name bcus we have user_id
df_cust=df_cust.drop(columns=['First_name', 'Last_name'])
```

### mixed data types

```python
#check for mixed data columns
for col in df_cust.columns.tolist():
    weird=(df_cust[[col]].map(type) !=df_cust[[col]].iloc[0].apply(type)).any(axis=1) 
    if len(df_cust[weird])>0:
        print(col)
```
